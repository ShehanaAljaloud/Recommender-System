## Recommender System for Spoonacular Food Dataset 

### Problem Statment 

> Build recommender system to recommend the best 10 recipes to user based on (vegetarian, vegan, alcoholFree, calories)

Create column to see if the recipe is alcohol free or not 

```python
# bulid function if Alcohol/g column is 0 it returns True else return False 
def sett(x):
    if x == 0:
        return True
    else:
        return False

# initiate the new column and assign values to it 
df['alcoholFree'] = df['Alcohol/g'].apply(sett)
```

Take answers from user

```python
# ask user if he/she is vegetarian or not 
vegetarian = input("Are you vegetarian? ")
# ask user if he/she is vegan or not 
vegan = input("Are you vegan? ")
# ask user if he/she is drunk Alcohol or not 
alcohol = input("Do you drunk Alcohol? ")
# ask user about the maximum amount of calories that he/she can eats
h_calories = input("highest amount of calories ")
# ask user about the minimum amount of calories that he/she needs 
l_calories = input("lowest amount of calories ")


#  ['yes', 'True', 'true', 't','y'] is list of all possible positive answers
# ['no', 'False', 'false', 'f','n'] is list of all possible negative answers 

#change vegetarian type from str to bool
if vegetarian in ['yes', 'True', 'true', 't','y']:
    vegetarian = True 
elif vegetarian in ['no', 'False', 'false', 'f','n']:
    vegetarian = False   
    
#change vegan type from str to bool     
if vegan in ['yes', 'True', 'true', 't','y']:
    vegan= True 
elif vegan in ['no', 'False', 'false', 'f','n']:
    vegan= False  
    
#change alcohol type from str to bool   
if alcohol in ['yes', 'True', 'true', 't','y']:
    alcohol= False 
elif alcohol in ['no', 'False', 'false', 'f','n']:
    alcohol= True
    
#change calories type from str to int
h_calories = int(h_calories)
l_calories = int(l_calories)
```

New dataset based on user's answers

```python
# create dataset contiens all similar items 
foods =df[(df['vegan']== vegan)&
   (df['vegetarian']== vegetarian)&
   (df['alcoholFree']== alcohol)&
   # get all calories between l_calories and h_calories
   (df['calories']>= l_calories)&
   (df['calories']<= h_calories)
  ]
```

```python
# sort items based on spoonacular score
foods = foods.sort_values('spoonacularScore', ascending=False)
# take a look at the data
foods.head()
```

Recommend recipes to user 

```python
# recommend the best 10 recipes to user 
print ('The best 10 recipes for you are:')
print(list(foods['title'])[:10])
```
